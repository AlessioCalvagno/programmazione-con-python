## Esercizio 2 - Analisi di magazzino
Aprire il file shirts.csv e calcola le seguenti statistiche su di esso:
1. Numero totale di prodotti presenti.
1. Valore totale del magazzino (somma del prezzo di tutti i prodotti).
1. Prezzo medio dei prodotti.
1. Numero di prodotti per ogni colore.
1. Numero di prodotti per ogni taglia.

Salva le statistiche in un file json chiamato "shirts_stats.json".



```python
import csv
import json
file_csv = open("shirts.csv")
csv_reader = csv.DictReader(file_csv)
#print(csv_reader.fieldnames)
#for row in csv_reader:
#   print(row)

product_count = 0 #punto 1
price_total = 0.0 #punto 2
product_for_color = {} #punto 4
product_for_size = {} #punto 5

for item in csv_reader:
    product_count += 1
    price_total += float(item["prezzo"])
    #per il punto 4 e 5 salvo i conteggi per ogni categoria in una coppia
    #chiave-valore apposita. Questo if serve per inizializzare.
    if(not item["colore"] in product_for_color):
        product_for_color[item["colore"]] = 1
    else:
         product_for_color[item["colore"]] += 1
        
    if(not item["taglia"] in product_for_size):
        product_for_size[item["taglia"]] = 1
    else:
         product_for_size[item["taglia"]] += 1

price_mean = price_total/product_count #punto 3

print(f"Total product number: {product_count}")
print(f"Total price: {price_total:.2f}")
print(f"Mean price: {price_mean:.2f}")
print(product_for_color)
print(product_for_size)

#salvataggio in nuovo file
stats = {
    "product_count":product_count,
    "total_price":price_total,
    "mean_price":price_mean,
    "color_count":product_for_color,
    "size_count":product_for_size
}

file_json = open("shirts_stats.json","w")

json.dump(stats,file_json,indent=4)


file_csv.close()
file_json.close()
```
